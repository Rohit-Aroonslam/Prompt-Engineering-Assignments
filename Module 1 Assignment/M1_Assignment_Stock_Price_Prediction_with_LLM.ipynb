# Step 2 - Feature Engineering

### Explanation of Engineered Features and Their Usefulness for Stock Price Prediction:

1.  **Moving Averages (MA_5, MA_10, MA_50)**:
    *   **What they are:** Moving averages smooth out price data over a specified period, reducing noise and highlighting underlying trends. `MA_5` and `MA_10` represent short-term trends, while `MA_50` indicates a medium-term trend.
    *   **Usefulness:**
        *   **Trend Identification:** An upward-sloping MA suggests an uptrend, and a downward-sloping MA suggests a downtrend. Longer MAs confirm longer-term trends.
        *   **Support and Resistance:** MAs can act as dynamic support (a price level where buying interest is strong enough to stop further price decline) or resistance (a price level where selling interest is strong enough to stop further price appreciation).
        *   **Crossovers:** Crossovers between different moving averages (e.g., a short-term MA crossing above a long-term MA) are often used as buy or sell signals in technical analysis.

2.  **Daily Returns (`daily_returns`)**:
    *   **What it is:** This measures the percentage change in the stock's closing price from one trading day to the next.
    *   **Usefulness:**
        *   **Momentum Indicator:** Positive returns indicate upward momentum, while negative returns indicate downward momentum. This can be a key input for predicting price direction.
        *   **Target Variable:** Daily returns are frequently used as a target variable for classification models (e.g., predicting whether returns will be positive or negative).

3.  **10-day Rolling Volatility (`volatility_10`)**:
    *   **What it is:** This is the standard deviation of daily returns over the past 10 trading days. It quantifies the degree of variation of a trading price series over time.
    *   **Usefulness:**
        *   **Risk Assessment:** Higher volatility implies greater price fluctuations and thus higher risk associated with the stock. Lower volatility suggests more stable price movements.
        *   **Trading Strategy Adjustment:** Traders and investors use volatility to adjust their strategies. High volatility might lead to wider stop-loss orders or smaller position sizes, while low volatility might precede significant price movements (breakouts or breakdowns).

```python
import pandas as pd
import numpy as np

# 1. Print current column names
print("Current columns before processing:", data.columns.tolist())

# Create a dictionary to hold the new single-level column data
temp_single_level_data = {}

# Check if columns are MultiIndex (which is the case for 'data' from yfinance)
if isinstance(data.columns, pd.MultiIndex):
    for original_col_tuple in data.columns:
        # Assuming format like ('Close', 'AAPL'), ('Volume', 'AAPL')
        if isinstance(original_col_tuple, tuple) and len(original_col_tuple) > 0:
            base_name = original_col_tuple[0] # Get 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'
            # Prioritize 'Adj Close' to be the 'Close' column
            if base_name == 'Adj Close':
                temp_single_level_data['Close'] = data[original_col_tuple]
            elif base_name in ['Open', 'High', 'Low', 'Close', 'Volume']:
                # If 'Close' is already set by 'Adj Close', don't overwrite it with a simple 'Close'
                if base_name == 'Close' and 'Close' in temp_single_level_data:
                    continue
                temp_single_level_data[base_name] = data[original_col_tuple]
else:
    # Fallback if columns are already single-level but might need renaming (less likely for yfinance raw output)
    for col_name in data.columns:
        if col_name == 'Adj Close':
            temp_single_level_data['Close'] = data[col_name]
        elif col_name in ['Open', 'High', 'Low', 'Close', 'Volume']:
            if col_name == 'Close' and 'Close' in temp_single_level_data:
                continue
            temp_single_level_data[col_name] = data[col_name]

# Create a new DataFrame from the dictionary, preserving the index, and assign it back to 'data'
# This ensures we have a DataFrame with only the desired single-level columns.
data = pd.DataFrame(temp_single_level_data, index=data.index)

# Ensure the required columns exist and are in the desired order
required_base_cols_order = ['Open', 'High', 'Low', 'Close', 'Volume']
if not all(col in data.columns for col in required_base_cols_order):
    missing = [col for col in required_base_cols_order if col not in data.columns]
    raise ValueError(f"Missing required columns after initial processing: {missing}")

# Reorder columns to the specified order and remove any extra columns
data = data[required_base_cols_order].copy() # .copy() to avoid SettingWithCopyWarning later

# 4. Create the specified features directly in the 'data' DataFrame
# 5-Day Moving Average
data['MA_5'] = data['Close'].rolling(window=5).mean()
# 10-Day Moving Average
data['MA_10'] = data['Close'].rolling(window=10).mean()
# 50-Day Moving Average
data['MA_50'] = data['Close'].rolling(window=50).mean()
# Daily Returns: Percentage change in the closing price from the previous day
data['Daily_Returns'] = data['Close'].pct_change()
# 10-Day Rolling Volatility: Standard deviation of Daily_Returns over a 10-day window
data['Volatility_10'] = data['Daily_Returns'].rolling(window=10).std()

# 5. Remove rows containing missing values created by rolling calculations.
# These will typically be the first few rows of the DataFrame due to the rolling window operations.
data.dropna(inplace=True)

# 6. Print the updated column names
print("\nUpdated column names of the engineered DataFrame:", data.columns.tolist())
# 7. Display the first 10 rows
print("\nFirst 10 rows of the feature-engineered dataset:")
display(data.head(10))
```

# Step 3.1 – Linear Regression

### Interpretation of Linear Regression Model Performance:

The **Mean Squared Error (MSE)**, **Mean Absolute Error (MAE)**, and **Root Mean Squared Error (RMSE)** provide numerical assessments of the model's prediction accuracy. Lower values for these metrics indicate a more accurate model. MAE is particularly useful as it represents the average absolute difference between the predicted and actual values, expressed in the same units as the target variable (dollars in this case).

The **R² Score** (Coefficient of Determination) quantifies the proportion of the variance in the target variable (`Tomorrow_Close`) that is predictable from the independent variables (our features). An R² score close to 1 suggests that the model explains a large portion of the target variable's variability, indicating a strong fit. Conversely, an R² close to 0 or negative implies that the model does not explain the target variable's variance well, or performs worse than simply predicting the mean. For financial time series like stock prices, achieving a high R² is often challenging due to market efficiency and inherent randomness, making even small positive R² values potentially significant.

### Explanation of Engineered Features and Their Usefulness for Stock Price Prediction:

*   **MA_5, MA_10, MA_50 (Moving Averages)**: These indicators smooth out price data over 5, 10, and 50 periods, respectively, to identify trends. Short-term MAs (5, 10) show immediate price momentum, while longer-term MAs (50) indicate broader trends. They help identify support/resistance levels and potential trend reversals through crossovers. For stock price prediction, they can serve as leading or lagging indicators of price direction and stability.
*   **Daily_Returns**: Represents the percentage change in the closing price from the previous day. It's a key measure of a stock's daily performance and momentum, often used as a direct or transformed target variable for prediction, or as an input feature to capture short-term price movements.
*   **Volatility_10 (10-Day Rolling Standard Deviation of Daily_Returns)**: Measures the stock's price fluctuation (risk) over the last 10 days. Higher volatility suggests higher risk and larger price swings, while lower volatility indicates more stable prices. It's crucial for risk management, understanding market sentiment, and adapting trading strategies, as periods of high volatility often precede significant price moves.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 1. Print the heading
print("============================================================")
print("STEP 3.1 - LINEAR REGRESSION MODEL")
print("============================================================")

# Create a copy of the data DataFrame to avoid modifying the original and to ensure the latest state
model_data = data.copy()

# 2. Create a target column named Tomorrow_Close by shifting the Close price one day forward.
model_data['Tomorrow_Close'] = model_data['Close'].shift(-1)

# 3. Create a new DataFrame called model_data using: model_data = data.dropna().copy()
# This step is already implicitly done by creating 'model_data' and then dropping NaNs on it.
model_data.dropna(inplace=True)

# 4. Use these predictor variables:
features = [
    'Open',
    'High',
    'Low',
    'Close',
    'Volume',
    'MA_5',
    'MA_10',
    'MA_50',
    'Daily_Returns',
    'Volatility_10'
]

# Define X (features) and y (target)
X = model_data[features]
y = model_data['Tomorrow_Close']

# 5. Split the data chronologically
# 70% Training, 30% Testing
split_index = int(len(X) * 0.7)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"\nTraining data size: {len(X_train)} samples")
print(f"Testing data size: {len(X_test)} samples")

# 6. Train a Linear Regression model.
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# 7. Display Mean Squared Error, Mean Absolute Error, Root Mean Squared Error, R² Score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"\nMean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")

# 8. Display the metrics again in a pandas DataFrame.
metrics_df = pd.DataFrame({
    'Metric': ['MSE', 'MAE', 'RMSE', 'R2 Score'],
    'Value': [mse, mae, rmse, r2]
})
print("\nModel Performance Metrics:")
display(metrics_df)

# 9. Plot Actual vs Predicted Closing Prices using Matplotlib.
plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual Tomorrow Close Price', color='blue', alpha=0.01)
plt.plot(y_test.index, y_pred, label='Predicted Tomorrow Close Price', color='red', linestyle='--', alpha=0.01)
plt.title('Actual vs. Predicted Tomorrow Close Prices (Linear Regression)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# STEP 3.2 – Random Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Print the heading
print("============================================================")
print("STEP 3.2 - RANDOM FOREST CLASSIFICATION:")
print("============================================================")

# Make a copy of the data DataFrame to work with
# Ensure 'data' has the 'Close' column from the previous step where columns were correctly named.
model_data = data.copy()

# 2. Create a binary target column named Target where:
# 1 = Tomorrow's Close price is higher than today's Close price
# 0 = Otherwise
model_data['Tomorrow_Close'] = model_data['Close'].shift(-1)
model_data['Target'] = (model_data['Tomorrow_Close'] > model_data['Close']).astype(int)

# 3. Create model_data = data.dropna().copy()
# This step is already handled by creating model_data and then dropping NaNs on it.
model_data.dropna(inplace=True)

# 4. Use the following predictors
features = [
    'Open',
    'High',
    'Low',
    'Close',
    'Volume',
    'MA_5',
    'MA_10',
    'MA_50',
    'Daily_Returns',
    'Volatility_10'
]

# Define X (features) and y (target)
X = model_data[features]
y = model_data['Target']

# 5. Split the data chronologically
# 70% Training, 30% Testing
# Do not shuffle.
split_index = int(len(X) * 0.7)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"\nTraining data size: {len(X_train)} samples")
print(f"Testing data size: {len(X_test)} samples")

# 6. Train a Random Forest Classifier using 100 trees and random_state=42.
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# 7. Display Accuracy, Confusion Matrix, Classification Report
accuracy = accuracy_score(y_test, y_pred_rf)
conf_matrix = confusion_matrix(y_test, y_pred_rf)
class_report = classification_report(y_test, y_pred_rf)

print(f"\nAccuracy: {accuracy:.4f}")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)

# 8. Plot Feature Importance as a horizontal bar chart.
feature_importances = pd.Series(rf_model.feature_importances_, index=features)
feature_importances.sort_values(ascending=True, inplace=True)

plt.figure(figsize=(10, 6))
feature_importances.plot(kind='barh')
plt.title('Feature Importances for Random Forest Classifier')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

### Interpretation of Random Forest Classifier Performance:

The **Accuracy** score provides the overall proportion of correctly classified instances. While a higher accuracy is generally better, it can be misleading in imbalanced datasets. For stock price prediction, even a small improvement over 50% accuracy can be significant.

The **Confusion Matrix** shows the number of true positive (correctly predicted price increase), true negative (correctly predicted no price increase), false positive (predicted increase, but price decreased), and false negative (predicted decrease, but price increased) predictions. This helps understand where the model is making errors.

The **Classification Report** offers more detailed metrics including Precision, Recall, and F1-Score for each class (0 for no increase, 1 for increase).
*   **Precision** for class 1 tells us what proportion of predicted increases were actually increases.
*   **Recall** for class 1 tells us what proportion of actual increases were correctly identified by the model.
*   **F1-Score** is the harmonic mean of precision and recall.

The **Feature Importance Plot** highlights which features the Random Forest model considered most influential in making its predictions. Features with higher importance values had a greater impact on the model's decision-making process for predicting whether the stock price would increase or decrease the next day. This can provide valuable insights into the drivers of stock price movement.

```python
# 2. Apply the same feature engineering steps used previously
# Make a copy to avoid modifying the original 'new_data_raw' DataFrame during intermediate steps
df_temp_new = new_data_raw.copy()

# Initialize a dictionary to store our cleaned base columns
base_cols_data_new = {}

# Check if columns are MultiIndex (common with yfinance output)
if isinstance(df_temp_new.columns, pd.MultiIndex):
    for original_col_tuple in df_temp_new.columns:
        if isinstance(original_col_tuple, tuple):
            base_name = original_col_tuple[0]
            if base_name == 'Adj Close' and 'Close' not in base_cols_data_new:
                base_cols_data_new['Close'] = df_temp_new[original_col_tuple]
            elif base_name == 'Close' and 'Close' not in base_cols_data_new:
                base_cols_data_new['Close'] = df_temp_new[original_col_tuple]
            elif base_name == 'Open':
                base_cols_data_new['Open'] = df_temp_new[original_col_tuple]
            elif base_name == 'High':
                base_cols_data_new['High'] = df_temp_new[original_col_tuple]
            elif base_name == 'Low':
                base_cols_data_new['Low'] = df_temp_new[original_col_tuple]
            elif base_name == 'Volume':
                base_cols_data_new['Volume'] = df_temp_new[original_col_tuple]
else:
    for original_col in df_temp_new.columns:
        if original_col in ['Adj Close'] and 'Close' not in base_cols_data_new:
            base_cols_data_new['Close'] = df_temp_new[original_col]
        elif original_col in ['Close', 'c_l_o_s_e'] and 'Close' not in base_cols_data_new:
            base_cols_data_new['Close'] = df_temp_new[original_col]
        elif original_col in ['Open', 'o_p_e_n']:
            base_cols_data_new['Open'] = df_temp_new[original_col]
        elif original_col in ['High', 'h_i_g_h']:
            base_cols_data_new['High'] = df_temp_new[original_col]
        elif original_col in ['Low', 'l_o_w']:
            base_cols_data_new['Low'] = df_temp_new[original_col]
        elif original_col in ['Volume', 'v_o_l_u_m_e']:
            base_cols_data_new['Volume'] = df_temp_new[original_col]

# Convert the dictionary of series into a DataFrame, ensuring the index is preserved
df_features_new = pd.DataFrame(base_cols_data_new, index=df_temp_new.index)

# Define the exact required base column names
required_base_cols = ['Open', 'High', 'Low', 'Close', 'Volume']

# Verify that these columns exist before continuing.
if not all(col in df_features_new.columns for col in required_base_cols):
    missing_cols = [col for col in required_base_cols if col not in df_features_new.columns]
    raise ValueError(f"Missing required base columns after processing: {missing_cols}. Current columns: {df_features_new.columns.tolist()}")

# Create the specified features
df_features_new['MA_5'] = df_features_new['Close'].rolling(window=5).mean()
df_features_new['MA_10'] = df_features_new['Close'].rolling(window=10).mean()
df_features_new['MA_50'] = df_features_new['Close'].rolling(window=50).mean()
df_features_new['Daily_Returns'] = df_features_new['Close'].pct_change()
df_features_new['Volatility_10'] = df_features_new['Daily_Returns'].rolling(window=10).std()

# Create a binary target column named Target
df_features_new['Tomorrow_Close'] = df_features_new['Close'].shift(-1)
df_features_new['Target'] = (df_features_new['Tomorrow_Close'] > df_features_new['Close']).astype(int)

# Remove rows containing missing values created by rolling calculations and shifting
new_model_data = df_features_new.dropna().copy()

print("\nFeature-engineered new data for prediction:")
display(new_model_data.head())
```
